```python
import matplotlib.pyplot as plt
import numpy as np
import json

# Updated parameter array for x-axis
param_values = [0.1, 0.5, 0.8, 1, 1.2, 1.8, 3.0]

# Metrics to plot
metrics = [
    "validity",
    "sparsity_count",
    "diversity",
    "proximity",
    "avg_10nn_distance",
    "avg_10nn_class0_probability"
]

metric_titles = [
    "Validity (higher is better)",
    "Sparsity Count (number of changed features, lower is better)",
    "Diversity (higher is better)",
    "Proximity (lower is better)",
    "k-NN Distance to Distribution",
    "k-NN Probability of Class 0"
]

# Define model keys for each number of counterfactuals
cf_counts = [3, 6, 10, 16]

# Color scheme for different counterfactual counts
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
markers = ['o', 's', '^', 'd']  # Different markers for each counterfactual count

# Function to map parameter value to key format
def param_to_key_format(param_val, param_type):
    # Format key based on parameter type and value
    if param_val < 1:
        # Format for values < 1 (e.g., 0.1 -> "01")
        formatted_val = f"{int(param_val*10):02d}" if param_val > 0 else "00"
    else:
        # Format for values >= 1 (e.g., 1.5 -> "15", 10 -> "100")
        if param_val == int(param_val):
            formatted_val = f"{int(param_val)}" if param_val < 10 else f"{int(param_val)}0"
        else:
            formatted_val = f"{int(param_val*10)}"
    
    # Special case for 1.8
    if param_val == 1.8:
        formatted_val = "18"
    
    return f"{param_type}{formatted_val}"

# Function to extract data for plotting
def extract_metric_data(data, metric, param_type):
    result = {}
    
    for cf_count in cf_counts:
        # Initialize arrays
        means = []
        stds = []
        
        # Get vanilla model data (param = 1)
        vanilla_key = f"{cf_count}cf"
        if vanilla_key in data:
            vanilla_mean = data[vanilla_key][metric][1]  # avg value
            vanilla_std = data[vanilla_key][metric][3]   # std value
        else:
            print(f"Warning: {vanilla_key} not found in data")
            vanilla_mean = 0
            vanilla_std = 0
        
        # Process each parameter value
        for param_val in param_values:
            if param_val == 1:
                # Use vanilla model data for param=1
                means.append(vanilla_mean)
                stds.append(vanilla_std)
            else:
                # Generate key
                key_suffix = param_to_key_format(param_val, param_type)
                key = f"{cf_count}cf_{key_suffix}"
                
                # Some keys might have alternative formatting
                alt_key = None
                if key not in data:
                    # Try alternative key format for special cases
                    if param_val == 1.2:
                        alt_key = f"{cf_count}cf_{param_type}12"
                    elif param_val == 1.8:
                        alt_key = f"{cf_count}cf_{param_type}18"
                    elif param_val == 3.0:
                        alt_key = f"{cf_count}cf_{param_type}30"
                
                if key in data:
                    means.append(data[key][metric][1])  # avg value
                    stds.append(data[key][metric][3])   # std value
                elif alt_key and alt_key in data:
                    means.append(data[alt_key][metric][1])  # avg value
                    stds.append(data[alt_key][metric][3])   # std value
                else:
                    print(f"Warning: Neither {key} nor alternative found in data")
                    # Use previous value or 0 if first value
                    means.append(means[-1] if means else 0)
                    stds.append(stds[-1] if stds else 0)
        
        result[cf_count] = {
            'means': means,
            'stds': stds
        }
    
    return result

# Load the JSON data
try:
    with open('evaluation_with_knn_metrics.json', 'r') as f:
        data = json.load(f)
    print("Successfully loaded evaluation_with_knn_metrics.json")

# Create plots for each metric
for i, metric in enumerate(metrics):
    # Create figure with two subplots (one for proximity, one for diversity)
    plt.figure(figsize=(16, 8))
    
    # Extract data for proximity parameter
    p_data = extract_metric_data(data, metric, 'p')
    
    # Extract data for diversity parameter
    d_data = extract_metric_data(data, metric, 'd')
    
    # Plot 1: Proximity variation
    plt.subplot(1, 2, 1)
    
    for j, cf_count in enumerate(cf_counts):
        means = p_data[cf_count]['means']
        stds = p_data[cf_count]['stds']
        
        # Plot mean line with markers
        plt.plot(param_values, means, marker=markers[j], label=f'n={cf_count}', 
                 color=colors[j], linewidth=2, markersize=8)
        
        # Add error bars for standard deviation (except for validity)
        if metric != "validity":
            # Plot error bars as vertical lines with T-tips
            for x, y, std in zip(param_values, means, stds):
                plt.errorbar(x, y, yerr=std, fmt='none', color=colors[j], 
                           capsize=5, capthick=1, elinewidth=1, alpha=0.7)
    
    plt.title(f"{metric_titles[i]} vs Proximity Weight", fontsize=14)
    plt.xlabel("Proximity Weight", fontsize=12)
    plt.ylabel(metric_titles[i], fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=10)
    
    # Plot 2: Diversity variation
    plt.subplot(1, 2, 2)
    
    for j, cf_count in enumerate(cf_counts):
        means = d_data[cf_count]['means']
        stds = d_data[cf_count]['stds']
        
        # Plot mean line with markers
        plt.plot(param_values, means, marker=markers[j], label=f'n={cf_count}', 
                 color=colors[j], linewidth=2, markersize=8)
        
        # Add error bars for standard deviation (except for validity)
        if metric != "validity":
            # Plot error bars as vertical lines with T-tips
            for x, y, std in zip(param_values, means, stds):
                plt.errorbar(x, y, yerr=std, fmt='none', color=colors[j], 
                           capsize=5, capthick=1, elinewidth=1, alpha=0.7)
    
    plt.title(f"{metric_titles[i]} vs Diversity Weight", fontsize=14)
    plt.xlabel("Diversity Weight", fontsize=12)
    plt.ylabel(metric_titles[i], fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=10)
    
    plt.tight_layout()
    
    # Save with more appropriate filename (no dots in parameter type)
    filename = f"{metric}_plots.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved plot to {filename}")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import re
from pathlib import Path

def extract_file_info(filename):
    """Extract counterfactual count and parameter info from filename"""
    # Remove .parquet extension
    name = filename.replace('.parquet', '')
    
    # Extract number of counterfactuals (e.g., "3cf", "6cf", "10cf", "16cf")
    cf_match = re.search(r'(\d+)cf', name)
    if not cf_match:
        return None, None, None, None
    
    n_cf = int(cf_match.group(1))
    
    # Check for parameter type and value
    # Proximity parameters: p01, p05, p08, p12, p18, p30
    p_match = re.search(r'_p(\d+)', name)
    if p_match:
        param_value = int(p_match.group(1))
        # Convert to actual parameter value
        if param_value < 10:
            actual_value = param_value / 10.0  # p01 -> 0.1, p05 -> 0.5, p08 -> 0.8
        else:
            actual_value = param_value / 10.0   # p12 -> 1.2, p18 -> 1.8, p30 -> 3.0
        return n_cf, 'proximity', actual_value, name
    
    # Diversity parameters: d01, d05, d08, d12, d18, d30
    d_match = re.search(r'_d(\d+)', name)
    if d_match:
        param_value = int(d_match.group(1))
        # Convert to actual parameter value
        if param_value < 10:
            actual_value = param_value / 10.0  # d01 -> 0.1, d05 -> 0.5, d08 -> 0.8
        else:
            actual_value = param_value / 10.0   # d12 -> 1.2, d18 -> 1.8, d30 -> 3.0
        return n_cf, 'diversity', actual_value, name
    
    # Base files without parameters (vanilla)
    if name.endswith('cf'):
        return n_cf, 'vanilla', 1.0, name
    
    return None, None, None, None

def count_counterfactuals_in_file(filepath):
    """Count the number of counterfactuals in a parquet file"""
    try:
        df = pd.read_parquet(filepath)
        return len(df)
    except Exception as e:
        print(f"Error reading {filepath}: {e}")
        return 0

def get_reference_count(reference_file='data_eval_norm.parquet'):
    """Get the reference count from data_eval_norm.parquet"""
    try:
        if os.path.exists(reference_file):
            df = pd.read_parquet(reference_file)
            return len(df)
        else:
            print(f"Warning: {reference_file} not found. Using default value of 1000.")
            return 1000  # Default fallback value
    except Exception as e:
        print(f"Error reading reference file: {e}")
        return 1000

def main():
    # Get reference count
    reference_count = get_reference_count()
    print(f"Reference count from data_eval_norm.parquet: {reference_count}")
    
    # Get all parquet files in current directory
    parquet_files = [f for f in os.listdir('.') if f.endswith('.parquet')]
    
    # Data structures to store results
    proximity_data = {}  # {n_cf: {param_value: ratio}}
    diversity_data = {}   # {n_cf: {param_value: ratio}}
    
    # Process each file
    for filename in parquet_files:
        n_cf, param_type, param_value, clean_name = extract_file_info(filename)
        
        if n_cf is None:
            continue
            
        print(f"Processing {filename}: n_cf={n_cf}, type={param_type}, value={param_value}")
        
        # Count counterfactuals in file
        cf_count = count_counterfactuals_in_file(filename)
        
        # Calculate ratio: cf_count / (reference_count * n_cf)
        expected_count = reference_count * n_cf
        ratio = cf_count / expected_count if expected_count > 0 else 0
        
        print(f"  Counterfactuals: {cf_count}, Expected: {expected_count}, Ratio: {ratio:.3f}")
        
        # Store data based on parameter type
        if param_type == 'proximity':
            if n_cf not in proximity_data:
                proximity_data[n_cf] = {}
            proximity_data[n_cf][param_value] = ratio
        elif param_type == 'diversity':
            if n_cf not in diversity_data:
                diversity_data[n_cf] = {}
            diversity_data[n_cf][param_value] = ratio
        elif param_type == 'vanilla':
            # Add vanilla to both proximity and diversity at param_value = 1.0
            if n_cf not in proximity_data:
                proximity_data[n_cf] = {}
            if n_cf not in diversity_data:
                diversity_data[n_cf] = {}
            proximity_data[n_cf][1.0] = ratio
            diversity_data[n_cf][1.0] = ratio
    
    # Define parameter values for plotting
    param_values = [0.1, 0.5, 0.8, 1.0, 1.2, 1.8, 3.0]
    
    # Define colors and markers for different n_cf values
    cf_counts = [3, 6, 10, 16]
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
    markers = ['o', 's', '^', 'd']
    
    # Create plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    
    # Plot 1: Proximity
    ax1.set_title('Counterfactual Ratio vs Proximity Weight', fontsize=14)
    ax1.set_xlabel('Proximity Weight', fontsize=12)
    ax1.set_ylabel('CF Ratio (CF Count / (Reference  n))', fontsize=12)
    ax1.grid(True, alpha=0.3)
    
    for i, n_cf in enumerate(cf_counts):
        if n_cf in proximity_data:
            x_vals = []
            y_vals = []
            
            for param_val in param_values:
                if param_val in proximity_data[n_cf]:
                    x_vals.append(param_val)
                    y_vals.append(proximity_data[n_cf][param_val])
            
            if x_vals:  # Only plot if we have data
                ax1.plot(x_vals, y_vals, marker=markers[i], label=f'n={n_cf}', 
                        color=colors[i], linewidth=2, markersize=8)
    
    ax1.legend(fontsize=10)
    ax1.set_ylim(0, 1.1)  # Set y-axis from 0 to 1.1 since ratios should be between 0 and 1
    
    # Plot 2: Diversity
    ax2.set_title('Counterfactual Ratio vs Diversity Weight', fontsize=14)
    ax2.set_xlabel('Diversity Weight', fontsize=12)
    ax2.set_ylabel('CF Ratio (CF Count / (Reference  n))', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    for i, n_cf in enumerate(cf_counts):
        if n_cf in diversity_data:
            x_vals = []
            y_vals = []
            
            for param_val in param_values:
                if param_val in diversity_data[n_cf]:
                    x_vals.append(param_val)
                    y_vals.append(diversity_data[n_cf][param_val])
            
            if x_vals:  # Only plot if we have data
                ax2.plot(x_vals, y_vals, marker=markers[i], label=f'n={n_cf}', 
                        color=colors[i], linewidth=2, markersize=8)
    
    ax2.legend(fontsize=10)
    ax2.set_ylim(0, 1.1)  # Set y-axis from 0 to 1.1 since ratios should be between 0 and 1
    
    plt.tight_layout()
    plt.savefig('counterfactual_ratios.png', dpi=300, bbox_inches='tight')
    print("Saved plot to counterfactual_ratios.png")
    plt.show()
    
    # Print summary statistics
    print("\n=== SUMMARY ===")
    print("Proximity Data:")
    for n_cf in sorted(proximity_data.keys()):
        print(f"  n={n_cf}: {dict(sorted(proximity_data[n_cf].items()))}")
    
    print("\nDiversity Data:")
    for n_cf in sorted(diversity_data.keys()):
        print(f"  n={n_cf}: {dict(sorted(diversity_data[n_cf].items()))}")

if __name__ == "__main__":
    main()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Define the hinge_loss function
def hinge_loss(predictions, desired_class=0):
    logits = -torch.log(1 / predictions - 1)
    if desired_class == 0:
        return torch.max(torch.zeros_like(logits), 1 + logits)
    else:
        return torch.max(torch.zeros_like(logits), 1 - logits)

# Generate prediction values
pred_values = torch.linspace(0.01, 0.99, 500)
loss_class_0 = hinge_loss(pred_values, desired_class=0).detach().numpy()
loss_class_1 = hinge_loss(pred_values, desired_class=1).detach().numpy()

# Plot the hinge loss
plt.figure()
plt.plot(pred_values.numpy(), loss_class_0, label='Desired class = 0')
plt.axvline(0.5, color='gray', linestyle='--', label='Decision boundary (0.5)')
plt.xlabel('Prediction (sigmoid output)')
plt.ylabel('Hinge Loss')
plt.title('Hinge Loss vs. Prediction')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
from tqdm import tqdm

def compute_changed_features(
    original_data_path,
    sparse_counterfactuals_path,
    tolerance=1e-6,
    output_path='sparsity_analysis.parquet',
    plot_path='sparsity_cdf.png'
):
    """
    Compute the number of changed features for each sparse counterfactual
    and generate cumulative distribution analysis
    
    Args:
        original_data_path: Path to original normalized data (data_eval_norm.parquet)
        sparse_counterfactuals_path: Path to sparse counterfactuals
        tolerance: Numerical tolerance for feature comparison
        output_path: Path to save detailed analysis results
        plot_path: Path to save the CDF plot
        
    Returns:
        Tuple of (n_values, cdf_values, analysis_df)
    """
    print("Loading data...")
    
    # Load original data and sparse counterfactuals
    original_data = pd.read_parquet(original_data_path)
    sparse_cfs = pd.read_parquet(sparse_counterfactuals_path)
    
    # Extract feature names (excluding class and original_index columns)
    cf_columns = list(sparse_cfs.columns)
    feature_names = cf_columns[1:-1]  # Exclude first (class) and last (original_index) columns
    num_features = len(feature_names)
    
    print(f"Analyzing {len(sparse_cfs)} counterfactuals with {num_features} features...")
    
    # Initialize results
    changed_features_count = []
    detailed_results = []
    
    # Check the index ranges for diagnostics
    max_original_idx = sparse_cfs['original_index'].max()
    min_original_idx = sparse_cfs['original_index'].min()
    original_data_indices = set(original_data.index)
    
    print(f"Original data index range: {original_data.index.min()} to {original_data.index.max()}")
    print(f"Sparse CFs original_index range: {min_original_idx} to {max_original_idx}")
    print(f"Original data has {len(original_data)} rows with indices")
    
    # Check how many counterfactuals can be matched
    cf_indices = set(sparse_cfs['original_index'].astype(int))
    matchable_indices = cf_indices.intersection(original_data_indices)
    print(f"Counterfactuals reference {len(cf_indices)} unique indices")
    print(f"Matchable indices: {len(matchable_indices)}")
    
    if len(matchable_indices) < len(cf_indices):
        unmatched = cf_indices - original_data_indices
        print(f"WARNING: {len(unmatched)} indices cannot be matched")
        print(f"Sample unmatched indices: {sorted(list(unmatched))[:10]}")
    
    skipped_count = 0
    
    # Process each counterfactual
    for idx, row in tqdm(sparse_cfs.iterrows(), total=len(sparse_cfs)):
        # Get the original index (this is a DataFrame index, not positional)
        original_idx = int(row['original_index'])
        
        try:
            # Use .loc to access by index rather than .iloc for position
            original_sample = original_data.loc[original_idx]
        except KeyError:
            skipped_count += 1
            if skipped_count <= 5:  # Only print first few warnings
                print(f"Skipping CF {idx}: Cannot find original sample with index {original_idx}")
            continue
        
        # Extract feature values
        cf_features = row[feature_names].values
        original_features = original_sample[feature_names].values
        
        # Compute absolute differences
        differences = np.abs(cf_features - original_features)
        
        # Count features that are different (beyond tolerance)
        changed_mask = differences > tolerance
        n_changed = np.sum(changed_mask)
        
        changed_features_count.append(n_changed)
        
        # Store detailed results
        detailed_results.append({
            'cf_index': idx,
            'original_index': original_idx,
            'n_changed_features': n_changed,
            'n_identical_features': num_features - n_changed,
            'max_difference': np.max(differences),
            'mean_difference': np.mean(differences),
            'cf_class': row.iloc[0],  # First column is classification
        })
    
    print(f"\nProcessed {len(changed_features_count)} counterfactuals successfully")
    if skipped_count > 0:
        print(f"Skipped {skipped_count} counterfactuals due to missing original indices")
    
    # Convert to arrays
    n_values = np.array(changed_features_count)
    
    # Compute statistics
    print(f"\nChanged Features Statistics:")
    print(f"Min changed features: {np.min(n_values)}")
    print(f"Max changed features: {np.max(n_values)}")
    print(f"Mean changed features: {np.mean(n_values):.2f}")
    print(f"Median changed features: {np.median(n_values):.2f}")
    print(f"Standard deviation: {np.std(n_values):.2f}")
    
    # Also show identical features statistics
    n_identical = num_features - n_values
    print(f"\nIdentical Features Statistics:")
    print(f"Min identical features: {np.min(n_identical)}")
    print(f"Max identical features: {np.max(n_identical)}")
    print(f"Mean identical features: {np.mean(n_identical):.2f}")
    print(f"Median identical features: {np.median(n_identical):.2f}")
    
    # Compute cumulative distribution for changed features
    unique_n_values = np.unique(n_values)
    cdf_values = np.zeros_like(unique_n_values, dtype=float)
    
    total_samples = len(n_values)
    for i, n in enumerate(unique_n_values):
        cdf_values[i] = np.sum(n_values <= n) / total_samples
    
    # Create detailed analysis DataFrame
    analysis_df = pd.DataFrame(detailed_results)
    
    # Save detailed results
    analysis_df.to_parquet(output_path)
    print(f"Detailed analysis saved to {output_path}")
    
    # Create and save plot
    create_sparsity_plot(unique_n_values, cdf_values, n_values, plot_path)
    
    return unique_n_values, cdf_values, analysis_df

def create_sparsity_plot(n_values, cdf_values, all_n_values, plot_path):
    """
    Create a comprehensive analysis plot for changed features
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Cumulative Distribution Function
    ax1.step(n_values, cdf_values, where='post', linewidth=2, color='blue', alpha=0.8)
    ax1.fill_between(n_values, cdf_values, alpha=0.3, color='blue', step='post')
    ax1.set_xlabel('Number of Changed Features (n)', fontsize=12)
    ax1.set_ylabel('Cumulative Probability', fontsize=12)
    ax1.set_title('Cumulative Distribution of Changed Features', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(left=0)
    ax1.set_ylim(0, 1)
    
    # Add percentile lines
    percentiles = [0.25, 0.5, 0.75, 0.9, 0.95]
    for p in percentiles:
        n_at_percentile = np.percentile(all_n_values, p * 100)
        ax1.axhline(y=p, color='red', linestyle='--', alpha=0.6, linewidth=1)
        ax1.axvline(x=n_at_percentile, color='red', linestyle='--', alpha=0.6, linewidth=1)
        ax1.text(n_at_percentile + max(n_values) * 0.01, p + 0.02, 
                f'{int(p*100)}%: {n_at_percentile:.0f}', 
                fontsize=9, color='red')
    
    # Plot 2: Histogram
    ax2.hist(all_n_values, bins=min(50, len(n_values)), alpha=0.7, color='skyblue', 
             edgecolor='black', linewidth=0.5)
    ax2.set_xlabel('Number of Changed Features (n)', fontsize=12)
    ax2.set_ylabel('Frequency', fontsize=12)
    ax2.set_title('Distribution of Changed Features', fontsize=14, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    
    # Add statistics text
    total_features = np.max(all_n_values) + np.min(all_n_values)  # Rough estimate based on range
    stats_text = f"""Changed Features Stats:
Min: {np.min(all_n_values)}
Max: {np.max(all_n_values)}
Mean: {np.mean(all_n_values):.2f}
Median: {np.median(all_n_values):.2f}
Std: {np.std(all_n_values):.2f}

Sparsity: Lower is sparser
(fewer changed features)"""
    
    ax2.text(0.98, 0.98, stats_text, transform=ax2.transAxes, 
             fontsize=9, verticalalignment='top', horizontalalignment='right',
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    
    plt.tight_layout()
    plt.savefig(plot_path, dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"Plot saved to {plot_path}")

def analyze_sparsity_by_class(analysis_df):
    """
    Analyze changed features patterns by counterfactual class
    """
    print("\nChanged Features Analysis by Class:")
    print("=" * 50)
    
    for class_val in sorted(analysis_df['cf_class'].unique()):
        class_data = analysis_df[analysis_df['cf_class'] == class_val]
        n_changed = class_data['n_changed_features']
        n_identical = class_data['n_identical_features']
        
        print(f"\nClass {class_val} ({len(class_data)} samples):")
        print(f"  Changed features - Min: {np.min(n_changed)}, Max: {np.max(n_changed)}")
        print(f"  Changed features - Mean: {np.mean(n_changed):.2f}, Median: {np.median(n_changed):.2f}")
        print(f"  Identical features - Min: {np.min(n_identical)}, Max: {np.max(n_identical)}")
        print(f"  Identical features - Mean: {np.mean(n_identical):.2f}, Median: {np.median(n_identical):.2f}")

def get_most_sparse_examples(analysis_df, n_examples=10):
    """
    Get the most sparse counterfactuals (fewest changed features)
    """
    most_sparse = analysis_df.nsmallest(n_examples, 'n_changed_features')
    
    print(f"\nTop {n_examples} Most Sparse Counterfactuals (Fewest Changed Features):")
    print("=" * 70)
    for idx, row in most_sparse.iterrows():
        print(f"CF Index: {row['cf_index']}, "
              f"Original Index: {row['original_index']}, "
              f"Changed Features: {row['n_changed_features']}, "
              f"Identical Features: {row['n_identical_features']}, "
              f"Class: {row['cf_class']}")
    
    return most_sparse

# Main execution function
def run_sparsity_analysis(
    original_data_path='data_eval_norm.parquet',
    sparse_counterfactuals_path='sparse_counterfactuals.parquet',
    tolerance=1e-6
):
    """
    Run complete sparsity analysis
    """
    # Compute changed features and generate CDF
    n_values, cdf_values, analysis_df = compute_changed_features(
        original_data_path=original_data_path,
        sparse_counterfactuals_path=sparse_counterfactuals_path,
        tolerance=tolerance
    )
    
    # Additional analyses
    analyze_sparsity_by_class(analysis_df)
    most_sparse = get_most_sparse_examples(analysis_df)
    
    print(f"\nReturning:")
    print(f"- n_values: array of unique changed feature counts ({len(n_values)} values)")
    print(f"- cdf_values: corresponding cumulative probabilities")
    print(f"- analysis_df: detailed results DataFrame ({len(analysis_df)} rows)")
    print(f"Note: Results show CHANGED features. Lower values = more sparse counterfactuals.")
    
    return n_values, cdf_values, analysis_df

# Example usage:
# n_vals, cdf_vals, results_df = run_sparsity_analysis(
#     original_data_path='data_norm.parquet',  # Use the training data file
#     sparse_counterfactuals_path='sparse_3cf_grad_desc.parquet'
# )



# Example usage:
n_vals, cdf_vals, results_df = run_sparsity_analysis(
    original_data_path='data_eval_norm.parquet',
    sparse_counterfactuals_path='sparse_3cf_grad_desc.parquet'
)
```
